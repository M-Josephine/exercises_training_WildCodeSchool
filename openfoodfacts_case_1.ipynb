# Taux de remplissage des colonnes

```
def fill_rate(dataframe):
  for colonne in range(dataframe.shape[1]):
    perc = (100*dataframe.iloc[:,colonne].notna().sum()) / dataframe.shape[0]
    print('colonne {} : {}%'.format(colonne, round(perc,2)))

fill_rate(df)
```

# Quartiles des taux de remplissage

```
#on reprend le meme principe que la fonction precedente et on ajoute un compteur pour chaque quartile
def quartile(dataframe):
  Q1 = 0
  Q2 = 0
  Q3 = 0
  Q4 = 0
  for colonne in range(dataframe.shape[1]):
    perc = (100*dataframe.iloc[:,colonne].notna().sum()) / dataframe.shape[0]
    if perc >=0 and perc <=25:
      Q1 += 1
    elif perc >25 and perc <=50:
      Q2 += 1
    elif perc >50 and perc <=75:
      Q3 += 1 
    else:
      Q4 += 1
  print('{} colonnes appartiennent au quartile 1'.format(Q1))
  print('{} colonnes appartiennent au quartile 2'.format(Q2))
  print('{} colonnes appartiennent au quartile 3'.format(Q3))
  print('{} colonnes appartiennent au quartile 4'.format(Q4))

quartile(df)
```
